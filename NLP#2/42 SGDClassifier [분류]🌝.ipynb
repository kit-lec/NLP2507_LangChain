# SGDClassifier

### ë°ì´í„° ì¤€ë¹„

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
````

````python
base_path = r'/content/drive/MyDrive/KoreaIT (á„á…©á„…á…µá„‹á…¡it)/250715 ğŸ’šá„Œá…¡á„‹á…§á†«á„‹á…¥á„á…¥á„…á…µS4/[á„Œá…¡á„‹á…§á†«á„‹á…¥á„á…¥á„…á…µ]/dataset(NLP2507)'
````

````python
file_path = os.path.join(base_path, 'fish.csv')

fish_df = pd.read_csv(file_path)

fish_df
````

````python
# ì…ë ¥ë°ì´í„°ì™€ íƒ€ê²Ÿë°ì´í„° ë¶„ë¦¬

fish_input = fish_df[['Weight','Length','Diagonal','Height','Width']].to_numpy()
fish_target = fish_df['Species'].to_numpy()
````

````python
# í›ˆë ¨ / í…ŒìŠ¤íŠ¸ ì„¸íŠ¸ ë‚˜ëˆ„ê¸°
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(
    fish_input, fish_target, random_state=42)
````

````python
# ì „ì²˜ë¦¬ (í‘œì¤€í™”)
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
````

###  ëª¨ë¸ í•™ìŠµ (epoch)

````python
# ì‚¬ì´í‚·ëŸ°ì—ì„œ í™•ë¥ ì ê²½ì‚¬í•˜ê°•ë²•ì„ ì œê³µí•˜ëŠ” ëŒ€í‘œì ì¸ ë¶„ë¥˜ìš© í´ë˜ìŠ¤ SGDClassifier
from sklearn.linear_model import SGDClassifier
````

https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.SGDClassifier.html
```python
class sklearn.linear_model.SGDClassifier(
  loss='hinge',  # ì†ì‹¤í•¨ìˆ˜ ì¢…ë¥˜
  *,
  penalty='l2',
  alpha=0.0001,
  l1_ratio=0.15,
  fit_intercept=True,
  max_iter=1000,  # ìˆ˜í–‰í•  epoch íšŸìˆ˜
  tol=0.001, # ì¼ì • epoch ë™ì•ˆ ì„±ëŠ¥ì´ í–¥ìƒë˜ì§€ ì•Šìœ¼ë©´ ìë™ìœ¼ë¡œ ë©ˆì¶¤ , None ìœ¼ë¡œ ì§€ì •í•˜ë©´ ìë™ìœ¼ë¡œ ë©ˆì¶”ì§€ ì•ŠìŒ
  shuffle=True,
  verbose=0,
  epsilon=0.1,
  n_jobs=None,
  random_state=None,
  learning_rate='optimal',
  eta0=0.0,
  power_t=0.5,
  early_stopping=False,
  validation_fraction=0.1,
  n_iter_no_change=5,
  class_weight=None,
  warm_start=False,
  average=False)
```

````python
# SGDClassifier ê°ì²´ë¥¼ ë§Œë“¤ ë•Œ 2ê°œì˜ ë§¤ê°œë³€ìˆ˜ë¥¼ ì§€ì •
#  loss=  ì†ì‹¤í•¨ìˆ˜ ì¢…ë¥˜   'log_loss' - ë¡œì§€ìŠ¤í‹± ì†ì‹¤í•¨ìˆ˜
#  max_iter=  ìˆ˜í–‰í•  epoch íšŸìˆ˜.
````

````python
# â€» ì‚¬ì´í‚·ëŸ° 1.1.0 ë²„ì „ ì´í•˜ì¼ ê²½ìš° 'log_loss'ë¥¼ 'log'ë¡œ ë°”ê¾¸ì–´ ì£¼ì„¸ìš”.
sc = SGDClassifier(
    loss='log_loss',
    max_iter=10,
    random_state=42,
)
sc.fit(train_scaled, train_target)
````

````python
print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
````

````python
# í›ˆë ¨ì„¸íŠ¸, í…ŒìŠ¤íŠ¸ì„¸íŠ¸ ì—ì„œì˜ ì ìˆ˜ê°€ ëª¨ë‘ ë‚®ë‹¤.
# max_iter=10 ì´ ë¶€ì¡±í•˜ë‹¤ => ConvergenceWarning ê²½ê³ 
````

````python
# í™•ë¥ ì  ê²½ì‚¬ í•˜ê°•ë²•ì€ 'ì ì§„ì  í•™ìŠµ' ê°€ëŠ¥
# ëª¨ë¸ì„ ë‹¤ì‹œ ë§Œë“¤ì§€ ì•Šê³ ë„,  ì´ë¯¸ í›ˆë ¨ëœ ëª¨ë¸ì— ì¶”ê°€ë¡œ í›ˆë ¨ì´ ê°€ëŠ¥í•˜ë‹¤

# ì‚¬ì´í‚·ëŸ¿ì—ì„œ ëª¨ë¸ì„ ì¶”ê°€ë¡œ ì´ì–´ì„œ í›ˆë ¨ => partial_fit() ì‚¬ìš©
#  fit() ê³¼ ì‚¬ìš©ë²•ì€ ê°™ìœ¼ë‚˜, í˜¸ì¶œí• ë•Œë§ˆë‹¤ 1 epoch ì”© ì´ì–´ì„œ í›ˆë ¨í•œë‹¤
````

````python
# partial_fit() ì„ í˜¸ì¶œí•˜ê³  ë‹¤ì‹œ ì ìˆ˜ í™•ì¸í•´ë³´ì

sc.partial_fit(train_scaled, train_target)

print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
````

````python
# ì—í¬í¬ ë¥¼ í•œë²ˆë” ì‹¤í–‰í•˜ë‹ˆ -> ì •í™•ë„ í–¥ìƒ.
# ì–¼ë§ˆë‹¤ ë” í›ˆë ¨í•˜ë©´ ì¢‹ì„ê¹Œ?
````

````python
# ë¹„ë¡ í•™ìŠµì— ëª¨ë“  í›ˆë ¨ì„¸íŠ¸ ë°ì´í„°ë¥¼ ì „ë‹¬í–ˆì§€ë§Œ,
# SGDClassifier ëŠ” ì „ë‹¬ë°›ì€ í›ˆë ¨ì„¸íŠ¸ì—ì„œ '1ê°œì”©' ìƒ˜í”Œì„ êº¼ì• ì–´ ê²½ì‚¬ í•˜ê°•ë²• ìˆ˜í–‰.
# SGDClassifier ëŠ” 'ë¯¸ë‹ˆë°°ì¹˜ / ë°°ì¹˜ ê²½ì‚¬í•˜ê°•ë²•' ì œê³µ ì•ˆí•¨
````

# ì—í¬í¬ì˜ ê³¼ëŒ€ / ê³¼ì†Œì í•©

````python
# 'ë„ˆë¬´ ì ì€ ì—í¬í¬ ' -> í›ˆë ¨ì„¸íŠ¸ ì™€ í…ŒìŠ¤íŠ¸ì„¸íŠ¸ì— ëª¨ë‘ ì˜ ë§ì§€ ì•ŠëŠ” underfit ê°€ëŠ¥ì„± ë†’ë‹¤
# 'ë„ˆë¬´ ë§ì€ ì—í¬í¬ ' -> í›ˆë ¨ì„¸íŠ¸ì— ë„ˆë¬´ ì˜ë§ê³  í…ŒìŠ¤íŠ¸ì„¸íŠ¸ì—ëŠ” ì˜¤íˆë ¤ ì ìˆ˜ ë‚˜ìœ overfit ê°€ëŠ¥ì„± ë†’ë‹¤
````

````python
# í›ˆë ¨ ì„¸íŠ¸ ì ìˆ˜ëŠ” ì—í¬í¬ê°€ ì§„í–‰ë ìˆ˜ë¡ ê¾¸ì¤€íˆ ì¦ê°€í•˜ì§€ë§Œ í…ŒìŠ¤íŠ¸ ì„¸íŠ¸ ì ìˆ˜ëŠ” ì–´ëŠ ìˆœê°„ ê°ì†Œí•˜ê¸° ì‹œì‘í•©ë‹ˆë‹¤.
# ë°”ë¡œ ì´ ì§€ì ì´ ëª¨ë¸ì´ ê³¼ëŒ€ ì í•©ë˜ê¸° ì‹œì‘í•˜ëŠ” ê³³ì…ë‹ˆë‹¤.
# ê³¼ëŒ€ ì í•©ì´ ì‹œì‘í•˜ê¸° ì „ì— í›ˆë ¨ì„ ë©ˆì¶”ëŠ” ê²ƒì„ ì¡°ê¸° ì¢…ë£Œ (early stopping) ì´ë¼ í•©ë‹ˆë‹¤.
````

````python
# â€»ì‚¬ì´í‚·ëŸ° 1.1 ë²„ì „ì—ì„œëŠ” SGDClassifierì˜ loss ë§¤ê°œë³€ìˆ˜ ì¤‘
#   ë¡œì§€ìŠ¤í‹± ì†ì‹¤ì„ ì˜ë¯¸í•˜ëŠ” 'log'ê°€ 'log_loss'ë¡œ ë°”ë€ë‹¤ëŠ” ê²½ê³ ê°€ ë°œìƒí•©ë‹ˆë‹¤.
#   ì‚¬ì´í‚·ëŸ° 1.1 ì´ìƒì„ ì‚¬ìš©í•˜ëŠ” ê²½ìš° loss='log'ë¥¼ loss='log_loss'ë¡œ ë³€ê²½í•˜ì„¸ìš”.

sc = SGDClassifier(loss='log_loss', random_state=42)
````

````python
# classes
classes = np.unique(train_target)
classes
````

````python
# 300ë²ˆì˜ ì—í¬í¬ ìˆ˜í–‰ => í›ˆë ¨ì„¸íŠ¸ì™€ í…ŒìŠ¤íŠ¸ì„¸íŠ¸ ì ìˆ˜ ê¸°ë¡

train_score = []
test_score = []
````

````python
for _ in range(0, 300):
  sc.partial_fit(train_scaled, train_target, classes=classes)

  train_score.append(sc.score(train_scaled, train_target))
  test_score.append(sc.score(test_scaled, test_target))
````

````python
# 300ë²ˆ ì˜ ì—í¬í¬ ë™ì•ˆ ê¸°ë¡í•œ í›ˆë ¨ì„¸íŠ¸ì™€ í…ŒìŠ¤íŠ¸ ì„¸íŠ¸ì˜ ì ìˆ˜ë¥¼ ì‹œê°í™”

plt.plot(train_score)
plt.plot(test_score)

plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.show()
````

````python
# SGDClassifier ì˜ ë°˜ë³µ íšŸìˆ˜(max_iter)ë¥¼ 100ì— ë§ì¶”ê³  ëª¨ë¸ì„ ë‹¤ì‹œ í›ˆë ¨í•˜ê³  ì ìˆ˜ í™•ì¸í•´ë³´ì

# SGDClassifier ëŠ” ì¼ì • ì—í¬í¬ ë™ì•ˆ ì„±ëŠ¥ì´ í–¥ìƒë˜ì§€ ì•Šìœ¼ë©´ ë” í›ˆë ¨í•˜ì§€ ì•Šê³  ìë™ìœ¼ë¡œ ë©ˆì¶¥ë‹ˆë‹¤.
# tol= ë§¤ê°œë³€ìˆ˜ëŠ” í–¥ìƒë  ìµœì†Ÿê°’ì„ ì§€ì •í•©ë‹ˆë‹¤. (tolerence)  (ê¸°ë³¸ê°’ 0.001)
#     None ìœ¼ë¡œ ì§€ì •í•˜ë©´ ìë™ìœ¼ë¡œ ë©ˆì¶”ì§€ ì•Šê³  max_iter ë§Œí¼ ë¬´ì¡°ê±´ ë°˜ë³µí•©ë‹ˆë‹¤


sc = SGDClassifier(loss='log_loss', max_iter=100, tol=None,  random_state=42)
sc.fit(train_scaled, train_target)

print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
````

````python
# íŒì§€ ì†ì‹¤í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•˜ì—¬ ê°™ì€ ë°˜ë³µ íšŸìˆ˜ë™ì•ˆ ëª¨ë¸ì„ í›ˆë ¨í•´ë³´ê² ìŠµë‹ˆë‹¤.

sc = SGDClassifier(loss='hinge', max_iter=100, tol=None, random_state=42)
sc.fit(train_scaled, train_target)

print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
````
