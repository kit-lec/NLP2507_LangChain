### re ì‚¬ìš©

````python
import re

def tokenize_sequence(text, filters=r"[!\"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n]", lower=True, split=" "):
  if lower:
    text = text.lower()

  text = re.sub(filters, "", text)
  return [word for word in text.split(split) if word]

````

````python
print('ë‹¨ì–´í† í°í™”4:', tokenize_sequence(sentence))
````

````python
import nltk
nltk.download('stopwords')
nltk.download('punkt_tab')
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
````

````python
raw_text = "A barber is a person. a barber is good person. a barber is huge person. he Knew A Secret! The Secret He Kept is huge secret. Huge secret. His barber kept his word. a barber kept his word. His barber kept his secret. But keeping and keeping such a huge secret to himself was driving the barber crazy. the barber went up a huge mountain."
````

````python
sentences = sent_tokenize(raw_text)
sentences
````

````python
vocab = {}  # ê° ë‹¨ì–´ë³„ ë¹ˆë„ìˆ˜
preprocessed_sentences = []
stop_words = set(stopwords.words('english'))

for sentence in sentences:
  # ê° ë¬¸ì¥ë§ˆë‹¤ ë‹¨ì–´ í† í°í™”
  tokenized_sentence = word_tokenize(sentence)

  result = []  # ì „ì²˜ë¦¬ëœ ë¬¸ì¥ì˜ ë‹¨ì–´ë“¤

  for word in tokenized_sentence:
    word = word.lower()  # ëª¨ë“  ë‹¨ì–´ë¥¼ ì†Œë¬¸ìí™” í•˜ì—¬ ë‹¨ì–´ì˜ ê°œìˆ˜ë¥¼ ì¤„ì¸ë‹¤.
    if word not in stop_words:  # ë‹¨ì–´ í† í°ì¤‘ ë¶ˆìš©ì–´ëŠ” ì œê±°
      if len(word) > 2:    # ë‹¨ì–´ ê¸¸ì´ê°€ 2 ì´í•˜ì¸ ê²½ìš° ì œê±°.
        result.append(word)
        # ë¹ˆë„ìˆ˜ ì²´í¬
        if word not in vocab:
          vocab[word] = 0
        vocab[word] += 1


  preprocessed_sentences.append(result)


preprocessed_sentences
````

````python
# ë‹¨ì–´ë¹ˆë„ìˆ˜ (ë‹¨ì–´ì§‘í•©)
print(vocab)
vocab
````

````python
vocab['barber']
````

````python
# ë¹ˆë„ìˆ˜ê°€ ë†’ì€ ìˆœì„œëŒ€ë¡œ ì •ë ¬

vocab_sorted = sorted(vocab.items(), key=lambda x:x[1], reverse=True)
print(vocab_sorted)

vocab_sorted
````

````python
# ë†’ì€ ë¹ˆë„ìˆ˜ë¥¼ ê°€ì§„ ë‹¨ì–´ì— ë‚®ì€ ì •ìˆ˜ë¥¼ ë¶€ì—¬.  ì •ìˆ˜ 1ë¶€í„° ë¶€ì—¬.

word_to_index = {}

idx = 0
for word, freq in vocab_sorted:
  if freq > 1: # ë¹ˆë„ìˆ˜ê°€ ì‘ì€ ë‹¨ì–´ëŠ” ì œì™¸
    idx += 1
    word_to_index[word] = idx

print(word_to_index)
word_to_index
````

````python
# ìœ„ ê²°ê³¼ì—ì„œ ìƒìœ„ 5ê°œì˜ ë‹¨ì–´ë§Œ ì‚¬ìš©í•´ë³´ì
````

````python
vocab_size = 5

words_frequency = [  # ì¸ë±ìŠ¤ê°€ 5 ì´ˆê³¼ì¸ ë‹¨ì–´ ì œê±°
    word
    for word, idx in word_to_index.items()
    if idx > vocab_size
]

for w in words_frequency:
  del word_to_index[w]   # í•´ë‹¹ ë‹¨ì–´ì— ëŒ€í•œ ì¸ë±ìŠ¤ ì •ë³´ ì‚­ì œ

print(word_to_index)

word_to_index
````

#### OOV (Out-Of-Vocabulary)

````python
# OOV ë¥¼ word_to_index  ì— ìƒˆë¡­ê²Œ ì¶”ê°€
# ë‹¨ì–´ì§‘í•©(ì‚¬ì „) ì— ì—†ëŠ” ë‹¨ì–´ëŠ” 'OOV' ë¡œ ì¸ë±ì‹± í• ê±°ë‹¤.

# ['barber', 'person'] -> [1, 5]
# ['barber', 'good', 'person'] -> [1, ??, 5]   <- ë‹¨ì–´ì§‘í•©ì— ì—†ëŠ” ë‹¨ì–´ëŠ” OOV ë¡œ ì¸ì½”ë”©
````

````python
word_to_index['OOV'] = len(word_to_index) + 1

print(word_to_index)
````

````python
# word_to_index  ë¥¼ ì‚¬ìš©í•˜ì—¬ sentences ì˜ ëª¨ë“  ë‹¨ì–´ë“¤ì„ ì •ìˆ˜ ì¸ì½”ë”©

encoded_sentences = []

for sentence in preprocessed_sentences:
  encoded_sentence = []
  for word in sentence:
    try:
      encoded_sentence.append(word_to_index[word])
    except KeyError:
      encoded_sentence.append(word_to_index['OOV'])
  encoded_sentences.append(encoded_sentence)

print(encoded_sentences)
encoded_sentences
````

````python
# ì›ë³¸ ë¬¸ì¥ê³¼ ì¸ì½”ë”©ëœ ë¬¸ì¥ì„ ê°™ì´ ì¶œë ¥
for e, s in zip(encoded_sentences, sentences):
  print(f'{str(e):25} {s}')
````

````python
def ont_hot_encoding(word, word_to_index):
  one_hot_vector = [0] * (len(word_to_index))  # ë‹¨ì–´ì§‘í•©ì˜ í¬ê¸°ë¥¼ ë²¡í„°ì˜ 'ì°¨ì›í¬ê¸°' ë¡œ í•˜ê³ 
  index = word_to_index[word]   #
  one_hot_vector[index] = 1  # í‘œí˜„í•˜ê³  ì‹¶ì€ ë‹¨ì–´ì˜ ì¸ë±ìŠ¤ì— '1' ê°’ì„ ë¶€ì—¬
  return one_hot_vector
````

````python
# 1ï¸âƒ£ ë‹¨ì–´ ì‚¬ì „ ì •ì˜
vocab = {"I": 0, "love": 1, "AI": 2, "hate": 3, "you": 4}
vocab_size = len(vocab)
embedding_dim = 5  # ê° ë‹¨ì–´ë¥¼ 5ì°¨ì› ë²¡í„°ë¡œ í‘œí˜„

# 2ï¸âƒ£ Embedding ë ˆì´ì–´ ìƒì„±
embedding = tf.keras.layers.Embedding(input_dim=vocab_size, output_dim=embedding_dim)

# 3ï¸âƒ£ ë¬¸ì¥ì„ ë‹¨ì–´ ì¸ë±ìŠ¤ë¡œ ë³€í™˜
sentence = ["I", "love", "AI"]
input_indices = tf.constant([vocab[word] for word in sentence])

print("ì…ë ¥ ë¬¸ì¥:", sentence)
print("ë‹¨ì–´ ì¸ë±ìŠ¤:", input_indices.numpy())

# 4ï¸âƒ£ ì„ë² ë”© ê²°ê³¼ ì–»ê¸°
embedded = embedding(input_indices)

print("\nì„ë² ë”© ê²°ê³¼ í…ì„œ shape:", embedded.shape)
print("ì„ë² ë”© ë²¡í„°:\n", embedded.numpy())
````
