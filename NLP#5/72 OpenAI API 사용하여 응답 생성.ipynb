```python
from openai import OpenAI
```

```python
client = OpenAI()   # OPENAI_API_KEY 환경변수가 있어야 한다.

# client = OpenAI(api_key="OpenAI 키를 입력하세요")
```

```python
messages = [
    {
        'role': 'system',
        'content': """
          너는 쇼핑몰 홈페이지에 올라온 질문에 대답하는 Q&A 챗봇이야.
          확정적인 답변을 하지 말고 제품 담당자가 정확한 답변을 하기 위해
          시간이 필요하다는 간단하고 친절한 답변을 생성해줘.
        """,
    },
    {
        'role': 'user',
        'content': """
          이 다이어리에 내년도 공휴일이 표시되어 있나요?
        """,
    }
]
```

```python
completion = client.chat.completions.create(
    model='gpt-4o-mini',
    messages=messages,
)

completion
```

```python
print(completion.choices[0].message.content)
```

```python
completion = client.chat.completions.create(
    model='gpt-4o-mini',
    messages=messages,
    temperature=1.8,  # LLM 출력의 무작위성 조절  (낮을수록 일관적, 결정적, 보수적 답변, 0.0~0.3)
)

completion
```

```python
completion = client.chat.completions.create(
    model='gpt-4o-mini',
    messages=messages,
    top_p=0.9, # LLM이 단어를 선택할때 확률이 높은 단어들만 골라서 그 안에서 무작위 선택하도록 제한
)

completion
```

## streaming=

```python
def stream_chat():
  completion = client.chat.completions.create(
      model='gpt-4o-mini',
      messages=messages,
      stream=True,  # streaming!
  )

  for chunk in completion:
    # delta 안에 새로 생성된 토큰이 담겨 있다.
    if hasattr(chunk.choices[0].delta, "content") and chunk.choices[0].delta.content:
      print(chunk.choices[0].delta.content, end="", flush=True)

stream_chat()
```

```python
# OpanAI Responses API 방식 (2024 이후)
# 신규 API에서는 client.responses.create() 를 사용하는 방식도 제공됩니다.

with client.responses.create(
    model="gpt-4.1",
    input="Python 언어데 대해 알려줘",
    stream=True,
) as stream:
    for event in stream:
        if event.type == "response.output_text.delta":
            print(event.delta, end="", flush=True)
```
