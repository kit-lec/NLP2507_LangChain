`````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
`````

`````python
base_path = r'/content/drive/MyDrive/KoreaIT (á„á…©á„…á…µá„‹á…¡it)/250715 ğŸ’šá„Œá…¡á„‹á…§á†«á„‹á…¥á„á…¥á„…á…µS4/[á„Œá…¡á„‹á…§á†«á„‹á…¥á„á…¥á„…á…µ]/dataset(NLP2507)'
`````

`````python
file_path = os.path.join(base_path, 'wine.csv')
wine_df = pd.read_csv(file_path)

wine_df
`````

`````python
data = wine_df[['alcohol', 'sugar', 'pH']].to_numpy()
target = wine_df['class'].to_numpy()
`````

`````python
from sklearn.model_selection import train_test_split

# train:test = 8:2 ë¡œ ë¶„ë¦¬
train_input, test_input, train_target, test_target = train_test_split(
    data, target, test_size=0.2, random_state=42)
`````

`````python
# cross_validate()
#  return_train_score=  ë¥¼ True ë¡œ ì§€ì •í•˜ë©´, ê²€ì¦ ì ìˆ˜ë¿ ì•„ë‹ˆë¼ í›ˆë ¨ì„¸íŠ¸ì— ëŒ€í•œ ì ìˆ˜ë¡œ ê°™ì´ ë¦¬í„´í•¨.
#                               => í›ˆë ¨ì„¸íŠ¸ì™€ ê²€ì¦ì„¸íŠ¸ ì ìˆ˜ë¥¼ ë¹„êµí•˜ì—¬ overfit íŒŒì•… ìš©ì´.
`````

`````python
from sklearn.model_selection import cross_validate
from sklearn.ensemble import RandomForestClassifier
`````

`````python
# cross validate ì‹¤í–‰
rf = RandomForestClassifier(n_jobs=-1, random_state=42)

scores = cross_validate(rf, train_input, train_target,
                        return_train_score=True,  # í›ˆë ¨ë°ì´í„° ì ìˆ˜ë„ ë¦¬í„´
                        n_jobs=-1)
`````

## íŠ¹ì„±ì¤‘ìš”ë„ permutation_importance()

`````python
# íŠ¹ì„±ì¤‘ìš”ë„ : permutation_importance() ì‚¬ìš©
#  íŠ¹ì„±ì„ í•˜ë‚˜ì”© ë¨ë¤í•˜ê²Œ ì„ì–´ì„œ ëª¨ë¸ì˜ ì„±ëŠ¥ì´ ë³€í™”í•˜ëŠ”ì§€ ê´€ì°°í•œ í›„ ì–´ë–¤ íŠ¹ì„±ì´ ì¤‘ìš”í•œì§€ ê³„ì‚°
#  n_repeats=  : ëœë¤í•˜ê²Œ ì„ì„ íšŸìˆ˜ ì§€ì • (ë””í´íŠ¸ 5)
`````

```python

  estimator,
  X,
  y,
  *,
  scoring=None,
  n_repeats=5, # ëœë¤í•˜ê²Œ ì„ì„ íšŸìˆ˜
```

`````python
# ë¨¼ì € íˆìŠ¤í† ê·¸ë¨ ê¸°ë°˜ ê·¸ë ˆë””ì–¸íŠ¸ ë¶€ìŠ¤íŒ… ëª¨ë¸ì„ í›ˆë ¨í•˜ê³  'í›ˆë ¨ ì„¸íŠ¸'ì—ì„œ íŠ¹ì„± ì¤‘ìš”ë„ ê³„ì‚°í•´ë³´ê² ìŠµë‹ˆë‹¤
# n_repeats= ë§¤ê°œë³€ìˆ˜ëŠ” ëœë¤í•˜ê²Œ ì„ì„ íšŸìˆ˜ë¥¼ ì§€ì •í•©ë‹ˆë‹¤.
#    ì—¬ê¸°ì„œëŠ” 10ìœ¼ë¡œ ì§€ì • í•˜ê² ìŠµë‹ˆë‹¤. (ë””í´íŠ¸ê°’ì€ 5 ì…ë‹ˆë‹¤)
`````
