````python
#  PyTorch ì™„ì „ ì¬í˜„ì„± ë³´ì¥ìš© ì‹œë“œ ì„¸íŒ… í•¨ìˆ˜


import os
import random
import numpy as np
import torch

def set_seed(seed: int = 42):
    """
    PyTorch ì™„ì „ ì¬í˜„ì„± ë³´ì¥ìš© ì‹œë“œ ì„¸íŒ… í•¨ìˆ˜

    Args:
        seed (int): ê³ ì •í•  ì‹œë“œ ê°’ (ê¸°ë³¸: 42)
    """
    # Python built-in random
    random.seed(seed)

    # í™˜ê²½ ë³€ìˆ˜
    os.environ["PYTHONHASHSEED"] = str(seed)

    # NumPy
    np.random.seed(seed)

    # PyTorch CPU
    torch.manual_seed(seed)

    # GPU (ë‹¨ì¼ ë° ë©€í‹°)
    torch.cuda.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

    # CuDNN ì„¤ì • - ì™„ì „ ê²°ì •ì  ëª¨ë“œ
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

    # PyTorch ì—°ì‚° ê²°ì •ì  ëª¨ë“œ (1.8+)
    torch.use_deterministic_algorithms(True)

    print(f"[INFO] Random seed set to {seed} (full reproducibility mode)")

# ì‚¬ìš© ì˜ˆì‹œ
# set_seed(1234)
````

## ê¸°ìš¸ê¸° (Gradient)

````python
x = torch.ones(3, 3, requires_grad=True)
x
````

````python
y = x + 5  # x ì— +5 ì—°ì‚°í•œ ê²°ê³¼ y
print(y)  # grad_fn=<AddBackward0>  xë¡œë¶€í„° ë§ì…ˆí•œ ê²°ê³¼ë¼ëŠ” ê¸°ë¡.
````

````python
# ìœ„ y ê°’ì— ë˜ ì—°ì‚°ì„ ìˆ˜í–‰í•´ë³´ì
z = y * y   # ì—°ì‚°
out = z.mean()  # ë˜ ì—°ì‚°

print(z)  # grad_fn=<MulBackward0>
print(out) # grad_fn=<MeanBackward0>
````

````python
x.grad  # .grad : ë¯¸ë¶„ê°’, ê¸°ìš¸ê¸°

# ì§€ê¸ˆì€ ê³„ì‚°ëœê²Œ ì—†ë‹¤.
# backward() í›„ì— ê³„ì‚°ëœ ê²ƒì„ ë³´ê²Œ ë  ê±°ë‹¤!
````

````python
out.backward()  # ì—­ì „íŒŒ ê³„ì‚°!  ì•Œì•„ì„œ ë¯¸ë¶„ì„ ìˆ˜í–‰í•´ ì¤€ë‹¤.
````

### .data ì†ì„±
- torch.Tensor ê°ì²´ì—ì„œ autograd(ìë™ ë¯¸ë¶„) ê´€ë ¨ ì •ë³´ ì—†ì´ ìˆœìˆ˜ ê°’ë§Œ ë‹´ì€ í…ì„œë¥¼ ë°˜í™˜.

- ë°˜í™˜ëœ í…ì„œëŠ” ì›ë³¸ í…ì„œì™€ ë©”ëª¨ë¦¬ë¥¼ ê³µìœ í•©ë‹ˆë‹¤.
â†’ ë”°ë¼ì„œ .dataë¥¼ í†µí•´ ê°’ì„ ë³€ê²½í•˜ë©´ ì›ë³¸ì—ë„ ì˜í–¥ì„ ë°›ìŒ

- ì£¼ì˜: autograd ê·¸ë˜í”„ë¥¼ ë¬´ì‹œí•©ë‹ˆë‹¤.
â†’ .dataë¥¼ ì‚¬ìš©í•´ ê°’ì„ ë°”ê¾¸ë©´, PyTorchì˜ ë¯¸ë¶„ ê¸°ë¡(ê·¸ë˜í”„)ì— ë°˜ì˜ë˜ì§€ ì•Šì•„ì„œ ê·¸ë˜ë””ì–¸íŠ¸ ê³„ì‚°ì´ ê¼¬ì¼ ìˆ˜ ìˆìŒ.

````python
torch.manual_seed(42)
x = torch.randn(3, requires_grad=True)
x
````

````python
y = x * 2

while y.data.norm() < 1000:
  y = y * 2

print(y)
````

````python
v = torch.tensor([0.1, 1.0, 0.0001], dtype=torch.float)

y.backward(v)  # v ê¸°ì¤€ìœ¼ë¡œ backward ê°€ ì´ë£¨ì–´ì§„ë‹¤.

x.grad  # x ì˜ ê¸°ìš¸ê¸°ê°€ ê³„ì‚°ì´ ë˜ì—ˆë‹¤!!!
````
