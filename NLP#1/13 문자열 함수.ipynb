# 등장 개수 세기
계수하기

```python
# 등장하는 알파벳의 개수 -> dict 결과 만들기
# Dict Comprehension 사용
# 대소문자 구분 하지 않기, 1개 이상 등장하는 알파벳만
# hint : 알파벳 리스트 , lower, upper, count

word = "Alice in wonderland"

# {'a': 2,
#  'c': 1,
#  'd': 2,
#  'e': 2,
#  'i': 2,
#  'l': 2,
#  'n': 3,
#  'o': 1,
#  'r': 1,
#  'w': 1}
```

```python
# 방법1 : for + dict

data = word.lower()

result = {}

for ch in data:
  if 'a' <= ch <= 'z':
    if not result.get(ch):  # 첫등장이면 1
      result[ch] = 1
    else:  # 이전에 최소 한번이라도 등장했다면
      result[ch] += 1

  print(f'{ch} -> {result}')


print(result)
```

```python
# 방법2 : get() + default값 사용

result = {}

for ch in word.lower():
  if ch.isalpha():
    result[ch] = result.get(ch, 0) + 1

print(result)
```

```python
# 방법3 : 문자열 메소드 + dict comprehension 사용

w = word.lower()
result = {ch:w.count(ch) for ch in w if ch.isalpha()}
result
```

```python
# 위 결과를 key 순서로 정렬.

data_list = list(result.items())

# sort(key=함수)
data_list.sort(key=lambda entry: entry[0])

dict(data_list)
```

```python
data_list = list(result.items())

data_list.sort(key=lambda entry: entry[1])

dict(data_list)
```

```python
sorted(result.items(), key=lambda entry: entry[0])
```

```python
sorted(result.items(), key=lambda entry: entry[1], reverse=True)
```

## 가장 긴 단어 찾기

```python
# *******************************
# 가장 긴 단어 찾기
#

data = [
    "I am a Student",        # ->  Student
    "That elephant is big",  # -> elephant
    "She loves cat very much", # -> loves
]

def longestWord(sentence):   # "I am a Student"
    max_word = "";  # 가장 길이가 긴 단어

    # 방법1
    # for word in sentence.strip().split():   # "I", "am", "a", "Student"
    #   if len(word) > len(max_word):
    #     max_word = word


    # "I", "am", "a", "Student"

    #        len()

    #  1 ,   2,   1 ,  6
    # ["Student", "am", "I", "a"]
    max_word = sorted(sentence.strip().split(), key=len, reverse=True)[0]

    return max_word;

[longestWord(d) for d in data]
# ↓ 결과 ['Student', 'elephant', 'loves']
```

## 문장에서 각 단어 첫글자만 대문자 만들기

```python
# *******************************
# 문장에서 각 단어 첫글자만 대문자 만들기
#

data = [
    "i am a PROGRAMMER",     # -> I Am A Programmer
    "THAT ELEPHANT IS BIG",  # -> That Elephant Is Big
]

def letterCapitalize(sentence):
    result = []

    for word in sentence.strip().lower().split():   # "i", "am", "a", "programmer"
      result.append(word[0].upper() + word[1:])  #  "P" + "rogrammer"

    return " ".join(result)

[letterCapitalize(d) for d in data]
# ↓ 결과 ['I Am A Programmer', 'That Elephant Is Big']
```
