# 요청할 URL 준비

```python
key = os.environ['SEOUL_API_KEY']
date = '20250801'
end_index = 10
```

```python
url = f'http://openapi.seoul.go.kr:8088/{key}/json/CardSubwayStatsNew/1/{end_index}/{date}'

print(url)
```

# requests 모듈
- http(웹 통신) 요청 - 응답

```python
import requests
```

```python
# get(url) : url 로 서버에 GET 방식으로 요청 -> Response 객체 리턴

response = requests.get(url)

response
```

```python
response.status_code
```

```python
response.headers['Content-Type']
```

```python
response.text
```

```python
# JSON reponse -> 파이썬 데이터 변환
data = response.json()
data
```

```python
import pprint

pprint.pprint(data)
```

```python
rows = data['CardSubwayStatsNew']['row']
len(rows)
```

```python
result = [
    {
        '호선명': row['SBWY_ROUT_LN_NM'],
        '역명': row['SBWY_STNS_NM'],
        '승차인원': row['GTON_TNOPE'],
        '하차인원': row['GTOFF_TNOPE'],
    }
    for row in rows
]

result
```

```python
import pandas as pd

pd.DataFrame(result)
```
