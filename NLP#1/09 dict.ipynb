# dict

```python
# 여러개의 데이터를 담는 집합 데이터 타입들..
#1. list   :  순서있다.  중복허용,  mutable
#2. tuple  :  순서있다,  중복허용,  immutable
#3. set : 순서없다,  중복허용안함
#4. dict : key, value 쌍으로 구성, 순서없다.
```

```python
# dict : 딕셔너리

# dictionary 데이터 타입은  key-value 쌍으로 저장되는 데이터 집합이다.

# {Key1:Value1, Key2:Value2, Key3:Value3 ...}
# "이름" = "홍길동", "생일" = "몇 월 몇 일"   과 같은 자료형 담음

# 기존의 list, tuple 등은 value 중심
# 그러나 이 또한 알고 보면 key-value 쌍으로 구성

# Python 3.7 부터는 dict 의 key 저장순서로 유지됨.
# key 는 중복 안된다.
```

# dict() 함수로 dict 생성

```python
# 빈 dict 생성
mySetting = {}

mySetting
```

```python
# 빈 dict 생성
mySetting = dict()

mySetting
```

```python
myConfig = dict(host='127.0.0.1', port=3306)  # keyword args 사용

myConfig
```

```python
# (key, value)  쌍을 담은 iterable 을 사용
pairs = [('a', 10), ('b', 20)]

d1 = dict(pairs)

d1
```

```python
# 위 두가지 방법 혼합 가능.

pairs = [('a', 10), ('b', 20)]

d1 = dict(pairs, host='127.0.0.1', port=3306)

d1
```

# key, value 의 타입
- **value** 는 어떠한 타입이라도 가능!
- **key** 는 hash 가능한 타입만 가능!   (ex: int, float, str, bool, tuple..)

```python
myDict = {
    1: 'haha',
    2: 'hoho',
    2: 'hehe',  # key 는 중복 불가.
    # [10, 20]: 3.14,  # list, set, dict 등은 dict 의 key 로 사용 불가
    "two": {
        3.14: "pi",
        "pi": 3.14,
    },
    False: [10, 20, 30],
    (1, 2): 'nice',  # tuple 이 key 로 사용될수 있다!  --> 나중에 Numpy, Pandas 등에서 활용됨.
}

myDict
```

```python
myDict[1]
```

```python
myDict['two']
```

```python
myDict['two'][3.14]
```

```python
myDict[False]
```

```python
myDict[False][1]
```

```python
myDict[(1, 2)]
```

```python
myDict[1, 2]
```

# dict.keys(),  dict.values(), dict.items()

```python
student = {
        "name":"최현진",
        "email":"choi@mail.com",
        "age": 23,
        "addr" : "서울"
        }

student
```

```python
student.keys()  # key 들로만 구성된 iterable 객체 리턴
```

```python
student.values()  # value 들로만 구성된 iterable 객체 리턴
```

```python
student.items()  # (key, value) 들로 구성된 iterable 객체 리턴
```

```python
# ※key-value 쌍을 entry 라고도 함.
```
