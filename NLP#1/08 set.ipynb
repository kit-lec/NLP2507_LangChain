```python
# 여러개의 데이터를 담는 집합 데이터 타입들..
#1. list   :  순서있다.  중복허용,  mutable
#2. tuple  :  순서있다,  중복허용,  immutable
#3. set : 순서없다,  중복허용안함
#4. dict : key, value 쌍으로 구성, 순서없다.
```

```python
# set
# 만드는 방법
# 1.set(iterable)  함수로 만들수도 있고
# 2. {  } 으로 만들수도 있다.
```

```python
animals = {"dog", "cat", 'dog', 'bird'}

animals  # 중복허용안함, 순서 없다!
```

```python
type(animals)
```

```python
len(animals)
```

```python
animals[0]
```

```python
animals = ["dog", "dog", 'cat', 'bird']
# 위 list 에서 중복된 데이터 제거하고 싶다면?
list(set(animals))
```

```python
# 빈 set 만들기
a = {}  # 이건 dict 다!
type(a)
```
