# while 순환문

```python
# while 순환문 구조,  조건식이 '참' 인 동안 수행문장을 반복수행

# while <조건식>:
#     <수행할 문장1>
#     <수행할 문장2>
#     <수행할 문장3>
#     ...
# else:
#      순환문 빠져나오기 전에 수행
```

```python
num = 0
while num < 10:
  print(num)  # 0  1  ...  9
  num += 1    # 0->1->2-> ...9 -> 10


print('종료', num)
```

```python
# 순환문에서 중요한 것은
# 1. 몇번 순환을 하는가?
# 2. 순환하는 동안 변수값의 변화 범위는?
# 3. 순환문 종료후 변수값은?

# 위 순환문의 경우
# 1. 총 10번 순환을 했고
# 2. 순환하는 동안 num 변수값은 0 부터 9 까지 변화
# 3. 순환문 종료후 num 값은 10
```

```python
# 20 ~ 10 출력

num = 20
while num >= 10:
  print(num, end=' ')
  num -= 3

print('종료', num)
```

# 중첩 순환문 (nested loop)

```python
# 중첩 순환문 (nested loop)
# 조건문 안에 조건문 블럭이 들어갈수 있듯이
# 순환문 안에도 얼마든지 순환문이 포함될수 있다..
# 조건문과 순환문 의 어떠한 조합도 가능하다
```

```python
# 구구단 출력 : 2단 ~ 9단
# 중첩된 while 문

'''
2 x 1 = 2
2 x 2 = 4
...
2 x 9 = 18
3 x 1 = 3
...
4 x 1 = 4
…
...
9 x 9 = 81
'''
None
```

```python
dan = 2

while dan <= 9:  # 2단 ~ 9단

  mul = 1
  while mul <= 9: # x1 ~ x9
    print(f'{dan} x {mul} = {dan * mul}')
    mul += 1

  dan += 1
```

# break

```python
# break
# 순환문 (while, for) 안 에서 순환문을 강제로 종료시키는 키워드

# break 는 감싸고 있는 가장 가까운 순환문을 종료합니다
```

```python
n = 1

while n <= 100:
  print(n)
  if n == 10:
    break
  n += 1

print('종료', n)
```

# continue

```python
# continue
# 순환문 처음으로 돌아가기
# 순환문은 종료하지 않되, 특정 조건만 skip 하는 경우 사용
```

```python
n = 1

while n <= 10:
  n += 1
  if n == 6: break
  print(n)

print('종료', n)
```

```python
n = 1

while n <= 10:
  n += 1
  if n == 6: continue
  print(n)

print('종료', n)
```

```python
# 정수를 계속해서 입력받다가, 0 이 입력되면
# 그때까지의 정수의 '합' 과 '평균' 을 출력

"""
[입력예]
10
20
30
0

[출력예]
합: 60
평균: 20.0

"""
total = 0  # 합
cnt = 0  # 개수


while True:
  n = int(input())
  if n == 0: break
  total += n
  cnt += 1

print('합:', total)
print('평균:%.1f' % (total / cnt))
```

```python
data = [11, 22, 33, 44]

# [22, 44, 66, 88]  <- 각 원소에 x 2 연산한 결과가 담긴 리스트 만들기

result = []

for d in data:
  result.append(d * 2)

print(result)


```

```python
data = [11, 22, 33, 44]

# [22, 44]  <-- 짝수만 걸러낸 리스트

result = []
for d in data:
  # if d % 2 == 0:
  #   result.append(d)
  (d % 2) or result.append(d)

print(result)
```

# zip()
- zip(iterable, iterable, ...)
- 매개변수 iterable 데이터들의 각 아이템들을 묶은 tuple들의 iterable 객체 생성

```python
list(zip([10, 20, 30], ['a', 'b', 'c']))
```

```python
for x in zip([10, 20, 30], ['a', 'b', 'c']):
  print(x)
```

```python
for x in zip([10, 20, 30], ['a', 'b', 'c'], "XYZ"):
  print(x)
```

```python
# 변수명 _  : 받아내긴 해야 하나 사용하지 않을 변수명으로 사용됨 (관례)

for num, ch, _ in zip([10, 20, 30], ['a', 'b', 'c'], "XYZ"):
  print(num, ch)
```

# Comprehension 구문

```python
# List Comprehension  -->  [ v for ...]  list 안에 포함된 for문  --> list 생성
# Dict Comprehension  -->  {k:v for ...}  dict 안에 포함된 for문 --> dict 생성
# Set Comprehension -->  {v for ...} set 안에 포함된 for문 --> set 생성
# Generator Expression -> (v for ...)  (v for ...)
```

## List Comprehension

```python
# List Comprehension 구문
# [표현식 for 항목 in 반복가능객체 (if 조건)]

# 주어진 리스트의 원소 데이터 들을 * 3 하여 새로운 리스트 작성하기
# ex) [1, 2, 3, 4] ==> [3, 6, 9, 12]

# 일단 지금까지 배운 방법으로 한다면

a = [1, 2, 3, 4]

result = []
for num in a:
  result.append(num * 3)

print(result)
```

```python
# list comprehension
[num * 3 for num in a]
```

```python
#  짝수에만 3을 곱하여 담고 싶다면 다음과 같이 "if 조건"을 사용할 수 있다
# [1, 2, 3, 4] --> [6, 12]

a = [1, 2, 3, 4]

# 실습!
[n * 3 for n in a if  n % 2 == 0 ]
```

```python
[
    n * 3
    for n in a
    if  n % 2 == 0
]
```

```python
# 아래와 같이 짝수에만 적용되게 하려면? 홀수는 그대로.
# [1, 2, 3, 4]
#  ↓  ↓  ↓  ↓
# [1, 6, 3, 12]   <- 2, 4 에만 x3 적용

a = [1, 2, 3, 4]

# 도전]

[
    num * 3 if num % 2 == 0 else num
    for num in a
]
```
