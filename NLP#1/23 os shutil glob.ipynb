# os모듈 : 디렉토리 관리

```python
import os
```

```python
# mkdir() : 디렉토리 한개 생성
savedir = 'tmpdir1'

if not os.path.exists(savedir):
  os.mkdir(savedir)
  print(savedir, '폴더 생성')
else:
  print(savedir, '이미 존재')
```

```python
# makedirs() : 중간 경로들 까지 모두 생성
savedir = r"tmpdir2/aaa/bbb/ccc"   # 경로는 raw-string 사용!

if not os.path.exists(savedir):
  os.makedirs(savedir)
  print(savedir, '폴더 생성')
else:
  print(savedir, '이미 존재')
```

```python
# listdir(path) 특정 디렉토리의 파일/디렉토리 목록 -> list

os.listdir('./sample_data')
```

```python
# getcwd() 현재 경로 (current working directory) -> str

os.getcwd()
```

```python
# os.path.join() 경로 문자열 생성

os.path.join(os.getcwd(), 'aaa', 'bbb', 'ccc/ddd/eee')
```

```python
# removedirs(path). 디렉토리 삭제

os.removedirs('tmpdir2')
# OSError: [Errno 39] Directory not empty: 'tmpdir2'
```

```python
os.removedirs(r'./tmpdir2/aaa/bbb/ccc')

# /tmpdir2 .. 밑 이하 디렉토리 완전 삭제
# ※ 이또한 os.path.exists() 로 확인한뒤 사용해야 한다
```

# glob 모듈
- 파일 시스템 상의 파일 경로를 패턴 매칭 방식으로 찾을 수 있게 해주는 파이썬 표준 라이브러리

```python
import glob
```

```python
# glob.glob(pattern) -> list[str]
# 지정한 패턴에 맞는 파일 경로 목록을 list로 리턴.

glob.glob('sample_data/*')
```

```python
glob.glob('sample_data/*.csv')
```

```python
# 주의! glob() 결과는 오름차순 정렬이 보장되지 않는다!

file_list = glob.glob('sample_data/*.csv')
file_list.sort()
file_list
```

# shutil
shell utility, 파일및 디렉토리 작업

```python
import shutil
```

```python
# copy(src, dst) : 파일 복사

shutil.copy('sample_data/README.md', 'My.md')
```

```python
# copytree(src, dst): 디렉토리 복사,  이하 디렉토리까지 복사

shutil.copytree('tmpdir1', 'tmpdir2')
```

```python
# os.removedirs('tmpdir2')

shutil.rmtree('tmpdir2')
```
