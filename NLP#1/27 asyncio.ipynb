```python
def sync_work(icon, seconds=0):
  print(f'{icon} {seconds}ì´ˆ ì‘ì—… ì‹œì‘')
  start = time.time()
  time.sleep(seconds)  # ì‹œê°„ì´ ê±¸ë¦¬ëŠ” ì‘ì—….
  end = time.time()
  print(f'{icon} ì‘ì—…ì¢…ë£Œ: {end - start:.3f}ì´ˆ ì†Œìš”')
```

# 2. ë¹„ë™ê¸° asyncio.sleep()

```python
async def async_work(icon, seconds=0):
  print(f'{icon} {seconds}ì´ˆ ì‘ì—… ì‹œì‘')
  start = time.time()
  await asyncio.sleep(seconds)  # ì‹œê°„ì´ ê±¸ë¦¬ëŠ” ì‘ì—… (ë¹„ë™ê¸°)
  end = time.time()
  print(f'{icon} ì‘ì—…ì¢…ë£Œ: {end - start:.3f}ì´ˆ ì†Œìš”')
```

```python
async def async_all():
  task1 = asyncio.create_task(async_work('ğŸ†', 3))
  task2 = asyncio.create_task(async_work('ğŸŠ', 1))
  task3 = asyncio.create_task(async_work('ğŸ¥¦', 2))

  await task1
  await task2
  await task3

# ì´ ëª‡ì´ˆì˜ ì‹œê°„ì´ ê±¸ë¦´ê²ƒì¸ê°€?
# í™”ë©´ì— ì¶œë ¥ë˜ëŠ” ìˆœì„œëŠ” ì–´ë–»ê²Œ ë ê²ƒì¸ê°€?
```

```python
start = time.time()
await async_all()
end = time.time()
print(f'>>> ì´ ê²½ê³¼ì‹œê°„: {end - start:.3f}ì´ˆ')
```
