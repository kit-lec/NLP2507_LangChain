```python
f = open("ìƒˆíŒŒì¼.txt", "r")

while True:
  line = f.readline()
  if not line: break    # ë” ì´ìƒ ì½ì„ ê²ƒì´ ì—†ìœ¼ë©´ None ë¦¬í„´
  print(line, end='')

f.close()
```

```python
# íŒŒì¼ê°ì²´ê°€ iterable í•˜ë‹¤

f = open("ìƒˆíŒŒì¼.txt", "r")

for line in f:
  print(line, end='')

f.close()
```
