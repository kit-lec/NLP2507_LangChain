| 목적     | 마크다운 문법                 | 결과 예시                                 |
| ------ | ----------------------- | ------------------------------------- |
| 제목     | `# 제목1` \~ `###### 제목6` | `# 제목` → <h1>제목</h1>                  |
| 굵은 글씨  | `**굵게**` 또는 `__굵게__`    | **굵게**                                |
| 기울임 글씨 | `*기울임*` 또는 `_기울임_`      | *기울임*                                 |
| 리스트    | `- 항목` 또는 `* 항목`        | - 항목                                  |
| 번호 리스트 | `1. 항목`                 | 1. 항목                                 |
| 링크     | `[링크텍스트](URL)`          | [구글](https://google.com)              |
| 이미지    | `![대체텍스트](이미지URL)`      | ![예시](https://via.placeholder.com/50) |
| 코드     | `` `코드` `` 또는 `코드`      | `코드`                                  |
| 인용문    | `> 인용문`                 | > 인용문                                 |
| 수평선    | `---` 또는 `***`          | ―――                                   |
