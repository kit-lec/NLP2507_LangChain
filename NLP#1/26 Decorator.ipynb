```python
def send_sms(func):  # func <= ëŒ€ìƒí•¨ìˆ˜  ê°€ ì „ë‹¨ëœë‹¤. (ì´ ê²½ìš° crawling)

  def inner_func():
    print('ğŸŸ¦ ëŒ€ìƒí•¨ìˆ˜ ì‹¤í–‰ì „')
    result = func()  # ëŒ€ìƒí•¨ìˆ˜ ì‹¤í–‰
    print('ğŸŒ SMS ë¥¼ ë³´ëƒ…ë‹ˆë‹¤')
    return result

  # decorator ëŠ” ë°˜ë“œì‹œ í•¨ìˆ˜ê°€ ë¦¬í„´ë˜ì•¼ í•˜ê³ , ê·¸ í•¨ìˆ˜ê°€ í˜¸ì¶œëœë‹¤
  # return lambda : print('ì§œì”')
  return inner_func

@send_sms
def crawling():  # crawling ì€ send_sms decorator ì˜ ëŒ€ìƒí•¨ìˆ˜(target function)
  print('í¬ë¡¤ë§ í•©ë‹ˆë‹¤')
  return 'í¬ë¡¤ë§ ê²°ê³¼'


crawling()  # crawling() í˜¸ì¶œí•˜ë©´ --> decorator ì¸ send_sms(crawling) ê°€ í˜¸ì¶œëœë‹¤!
```

# ë§¤ê°œë³€ìˆ˜ì™€ ë¦¬í„´ê°’ì„ ì²˜ë¦¬í•˜ëŠ” Decorator

```python
def trace(func):

  def wrapper(x, y):  # í˜¸ì¶œí•  ëŒ€ìƒí•¨ìˆ˜ì˜ ë§¤ê°œë³€ìˆ˜ë¥¼ ë°›ìŒ.
    r = func(x, y)
    print('{0}(a={1}, b={2}) -> {3}'.format(func.__name__, x, y, r))
    return r

  return wrapper


@trace
def add(a, b):   # ë§¤ê°œë³€ìˆ˜ ë‘ê°œ
  return a + b   # ë¦¬í„´ê°’!

print(add(10, 20))

print(add(-3, -5))

```

```python
# ê²½ê³¼ì‹œê°„ì¸¡ì • (elapsed time) í•˜ëŠ” Decorator ì‘ì„±í•˜ê¸°
import time
from datetime import timedelta

def elapsed_time(func):

  def count(loop, delay):
    print(f'â­•{func.__name__}({loop},{delay}) ì¸¡ì •ì‹œì‘')
    start = time.time()  # ì‹œì‘ì‹œê°„ ê¸°ë¡. í˜„ì¬ ì‹œê°„ ë¦¬í„´
    func(loop, delay)
    end = time.time() # ì¢…ë£Œì‹œê°„ ê¸°ë¡.
    print(f'âœ”ê²½ê³¼ì‹œê°„ {timedelta(seconds=end - start)}')

  return count


@elapsed_time
def countDown(loop, delay):  # countDown í•¨ìˆ˜ëŠ” ìˆ˜ì •í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤
    for i in range(loop, 0, -1):
        print(i)
        time.sleep(delay)

countDown(3, 2)
print()
countDown(5, 1)
print()

"""
[ì‹¤í–‰ê²°ê³¼]
countDown() ì¸¡ì •ì‹œì‘
3
2
1
ì¸¡ì •ì¢…ë£Œ 0:00:06.002019

countDown() ì¸¡ì •ì‹œì‘
5
4
3
2
1
ì¸¡ì •ì¢…ë£Œ 0:00:05.002277

"""
None
```
