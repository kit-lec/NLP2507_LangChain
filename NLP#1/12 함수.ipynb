# Argument Packing

## ê°€ë³€ë§¤ê°œë³€ìˆ˜ *args
### argument packing
ì…ë ¥ê°’ì´ ëª‡ê°œê°€ ì˜¬ë¥¸ì§€ ëª¨ë¥¸ë‹¤, ì„ì˜ì˜ ê°œìˆ˜ ì¸ì ë°›ê¸°

```python
# ê°€ë³€ë§¤ê°œë³€ìˆ˜(various arguments) í•¨ìˆ˜ êµ¬ë¬¸

# def í•¨ìˆ˜ì´ë¦„(*ë§¤ê°œë³€ìˆ˜):
#     <ìˆ˜í–‰í•  ë¬¸ì¥>
#     ...

# í•¨ìˆ˜ í˜¸ì¶œì‹œ ì „ë‹¬ëœ ë³µìˆ˜ê°œì˜ ë§¤ê°œë³€ìˆ˜ëŠ” í•˜ë‚˜ì˜ tuple ì˜ í˜•íƒœë¡œ ë¬¶ì—¬ì„œ(packing) ë‹¤ë£¨ì–´ì§„ë‹¤
# ê°€ë³€ ë§¤ê°œë³€ìˆ˜ë„ ì–´ë– í•œ ì´ë¦„ìœ¼ë¡œë„ ê°€ëŠ¥í•˜ê¸´ í•˜ë‚˜  ê´€ë¡€ì ìœ¼ë¡œ args ë¥¼ ë§ì´ ì‚¬ìš©í•œë‹¤
```

# ëŒë‹¤ (lambda)
'ì´ë¦„ì´ ì—†ëŠ”(ìµëª…)' inline í•¨ìˆ˜ ì •ì˜

êµ¬ë¬¸: `lambda [parameters]: expression`

```python
def square(x):
    return x ** 2

def absolute(x):
    return x * (1 if x > 0 else -1)
```

```python
absolute(-45)
```

```python
# ê°€ë ¹ ë‹¤ìŒê³¼ ê°™ì€ ê¸°ëŠ¥ì„ í•˜ëŠ” í•¨ìˆ˜ë“¤ë¥¼ ë§Œë“œë ¤ê³  í•œë‹¤

#      (ì…ë ¥)        (ë¦¬í„´)
# 1. [1, 2, 3] => [1, 4, 9]        <-- ì œê³±ì„ í•˜ëŠ” í•¨ìˆ˜

# 2. [-1, 2, -3] => [1, 2, 3]      <--  ì ˆëŒ€ê°’ì„ êµ¬í•˜ëŠ” í•¨ìˆ˜


# ì¦‰, ì§‘í•©ë°ì´í„°í˜•ì„ ì…ë ¥ ë°›ì•„ì„œ 'ê° ì›ì†Œ'ë“¤ì—ê²Œ 'ë¬´ì–¸ê°€ ì ìš©' í•œ ê²°ê³¼ë¥¼ ë§Œë“œëŠ” í•¨ìˆ˜!!
```

```python
def apply_func(numbers, func):  # í•¨ìˆ˜ë¥¼ ë§¤ê°œë³€ìˆ˜ ë°›ëŠ” í•¨ìˆ˜ ì •ì˜.
  return [func(number) for number in numbers]
```

```python
apply_func([3, -2, 7], square)
```

```python
apply_func([3, -2, 7], absolute)
```
