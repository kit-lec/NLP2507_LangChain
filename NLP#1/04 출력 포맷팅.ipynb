# 기본출력함수 print()

```python
# 파이썬의 가장 기본적인 출력 함수 print()

# print( 출력내용1, 출력내용2, 출력내용3, ...)
#  위와 같이 print() 함수의 parameter 로 출력할 내용들을 기입하면 된다

# print() 함수는 출력한후 자동적으로 "줄바꿈" 됩니다 (디폴트)
```

```python
print(10, 20, 30)
print()
print('hello')
print()
print(3.14)
```

```python
# print?   # 함수의 도움말
```

## end=
출력이 끝난뒤 마지막에 출력하는 문자열


```python
print(10, 20, 30)
print('hello')
print(3.14)
```

```python
print(10, 20, 30, sep='**')
```

# 이스케이프 문자 (escape character)

```python
# '문자열' 내에서 특수한 문자 출력할때 사용
# \와 조합하여 정의됨

# 많이 사용되는 이스케이프 코드
#    \n : 줄바꿈
#    \t : 탭
#    \\ : 역슬래시
#    \' : 홀따옴표
#    \" : 쌍따옴표
```

```python
print("Hello Python")
print("Hello \nPython")
```

```python
print("\t1\t2\t3")
print("\t10\t20\t30")
print("\t100\t200\t300")
```

```python
print("Hello\" \\Python")
```

# 문자열 포맷팅 (formatting)

## 방법1 % 연산자 사용

```python
# 서식지정자(format specifier) 를 포함한 문자열과,
#  각 '서식문자'에 대응하는 데이터들을 연결하여 문자열 완성

# 구문)
#    "서식문자포함문자열" % ( 데이터 튜플 )
```

```python
a = "Hello %s"

a
```

```python
a % ("파이썬")
```

```python
"내 이름은 %s, 나이는 %d 살" % ('홍길동', 20)
```

```python
"내 이름은 %s, 나이는 %d 살" % ('고영희', 5)
```

```python
# format specifier (서식 지정자)
#  https://docs.python.org/2/library/string.html

# %d   십진수 정수로 출력
# %f   실수로 출력
# %s   문자열로 출력
# %%   '%' 문자 자체
```

```python
"%d %f" % (10, 10)
```

```python
import math
```

```python
PI = math.pi
PI
```

```python
"%d %f %.1f %.2f %.3f" % (PI, PI, PI, PI, PI)
```

```python
a, b, c, d = 10, 100, 1000, 10000
print("%d:%d:%d:%d" % (a, b, c, d))
print("%5d:%5d:%5d:%5d" % (a, b, c, d))  # 5칸, 오른쪽 정렬
print("%-5d:%-5d:%-5d:%-5d" % (a, b, c, d))
```

## 방법2 format() 함수 사용

```python
myFmt3 = "My name is {}, I'm {} years old".format('킹콩', 100)

myFmt3
```

```python
myFmt3 = "My name is {name}, I'm {age} years old".format(name='킹콩', age=100)

myFmt3
```

```python
myFmt3 = "My name is {name}, I'm {age} years old".format(age=100, name='킹콩')

myFmt3
```

```python
print("PI = {}".format(PI))
print("PI = {:.3f}".format(PI))
```

## 방법3 f-string 사용

```python
lang = 'Python'
author = 'Guido van Rossum'
```

```python
'Language: {}, Author: {}'.format(lang, author)
```

```python
f'Language: {lang}, Author: {author}'
```

```python
f'원주율은 {PI} 입니다'
```

```python
f'원주율은 {PI:.2f} 입니다'
```
